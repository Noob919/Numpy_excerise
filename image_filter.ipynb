```python
plt.figure()
#read the image using matplotlib
plt.subplot(1,3,2)
img = plt.imread('dc_metro.png')
plt.imshow(img,cmap='hot')
 #function for the smoothing of the image
def smooth(img):
    avg_img =(    img[1:-1 ,1:-1]  # center
                + img[ :-2 ,1:-1]  # top
                + img[2:   ,1:-1]  # bottom
                + img[1:-1 , :-2]  # left
                + img[1:-1 ,2:  ]  # right
                ) / 5.0
    return avg_img

#Try to smooth the image using smooth function
plt.subplot(1,3,1)
img = plt.imread('dc_metro.png')
avg_img =  smooth(img)
plt.imshow(avg_img,cmap='hot')


plt.show()
```

```python
#see the differnce between the two image
plt.figure()
plt.gray()
#Plot the original Image
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('original')

#Plot the smoothed image 
plt.subplot(1,3,2)
plt.imshow(avg_img)
plt.title('Smoothed')

# And finally the difference between the two.
plt.subplot(1,3,3)
plt.imshow(img[1:-1,1:-1] - avg_img)
plt.title('difference')
```

```python
#see the differnce between the two image
plt.figure()
plt.gray()

#Plot the original Image
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('original')

#Plot the smoothed image 
plt.subplot(1,3,2)

for i in range(50):
    avg_img =  smooth(avg_img)

plt.imshow(avg_img,cmap= plt.cm.Greys_r)
plt.title('filtered 50 times')

 
```
